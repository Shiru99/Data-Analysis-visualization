### Drop

```python
fun01 = pd.read_csv("fun.csv")
fun01
```

```python
fun01.dropna()
#    or   
# fun01.dropna(how="any")                 # removes data containing any NAN
```

```python
fun01.dropna(how="all")                 # removes data containing all NAN
```

```python
fun01.dropna(subset=["Salary"])                # removes data of only paricular column containing all NAN
```

```python
fun01.head().dropna(axis=1)        # removes column completely containing NAN 
```

```python
fun01.fillna(0)                    # replaces all NAN with value 
```

```python
fun01["Salary"].fillna(0, inplace = True)       # only in salary column
# fun01["College"].fillna("No College", inplace =True)
fun01
```

```python
fun01.sort_values("College",ascending=False, na_position="last")
```

```python
fun01.sort_values(["Team","Name"], ascending=[True,False])
```

```python
fun01["SalaryRank"] = fun01["Salary"].rank(ascending=False).astype("int")
fun01
```

```python
mask = fun01["College"].isnull()
fun01[mask]
```

```python
mask = fun01["College"].notnull()
fun01[mask]
```

```python
fun01.rename(columns={"Salary":"Income","College":"School/University"}, inplace = True)
fun01
```

```python
fun01.columns = [column_name.replace("/"," or ") for column_name in fun01.columns]
fun01.head()
```

```python
mask = fun01["Name"] == "John Holland"
fun01.where(mask)
```

```python
fun01.query('Name == "John Holland"')
```

```python
fun01.query("Name in ['Jordan Mickey', 'John Holland','R.J. Hunter']")
```
