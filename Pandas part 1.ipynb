### 1. loc gets rows (or columns) with particular labels from the index.
#####            that is :
#####                         movieData.loc[0:100]  
#####                         prints until it finds index "100" it may be at 999th position, this will print upto that

### 2. iloc gets rows (or columns) at particular positions in the index (so it only takes integers).
#####            that is :
#####                        movieData.loc[0:100]  
#####                        prints till 100th position.

### 3. ix usually tries to behave like loc but falls back to behaving like iloc if a label is not present in the index.
#####            that is :
#####                         movieData.ix[0:100]  
#####                         LOC ==> prints until it finds index "100" it may be at 999th position, this will print upto that
#####                         if their is no "100" as index
#####                         it will
#####                         iLOC ==> prints till 100th position.
